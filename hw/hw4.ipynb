Problem 0

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

t = np.linspace(0,np.pi*100,np.int(1e5))

N = 10+np.sin(t)

bkgd = stats.norm.rvs(size=np.int(1e5))*np.sqrt(N)+N
```

Problem 1a

```python
fig, ax = plt.subplots(1,1)

ts, slice = t[:1000], bkgd[:1000]

ax.plot(ts,slice, '.')
```

1b

```python
def plot_slice(start, size):
    fig, ax = plt.subplots(1,1)
    
    ts, slice = t[start:(start+size)], bkgd[start:(start+size)]
    
    ax.plot(ts, slice, '.')
```

```python
plot_slice(500, 2000)
```

1c

```python
fig, ax = plt.subplots(1,1)

ts, slice = t[::100], bkgd[::100]

ax.plot(ts, slice, '.')
```

2a

```python
fig, ax = plt.subplots(1,1)

plt.hist2d(t, bkgd, 500)
plt.colorbar()
```

```python
plt.hist2d(t, bkgd, 1000)
plt.colorbar()
plt.show()
```

Above is a density of values from the background. Brighter points represent a greater density or number of values at that location.

3

```python
fold_t = [x % (2*np.pi) for x in t]

fig, ax = plt.subplots(1,1)

ax.plot(fold_t, bkgd, '.')
```

```python
plt.hist2d(fold_t, bkgd, 1000)
plt.colorbar()
plt.show()
```
